```python
import pandas as pd
import ta
import plotly.graph_objects as go
```

```python
data=pd.read_csv("C:/Users/erick/Downloads/aapl_1s_data (1).csv")
```

```python
def rsi_buy(data):
    rsi_indicator = ta.momentum.RSIIndicator(close=data['Close'], window=14)
    data['RSI'] = rsi_indicator.rsi()
    data["RSI_TF"]=False
    
    for i in range(len(data)):
        # Obtener el valor actual de RSI
        rsi_value = data['RSI'].iloc[i]

        # Si el valor de RSI es mayor que 50, establecer "True" en la columna "RSI_TF"
        if rsi_value > 50:
            data["RSI_TF"].iloc[i] = True

    return data
            
```

```python
rsi_buy(data)
```

```python
def macd_buy(data):
    macd_indicator = ta.trend.MACD(close=data['Close'], window_slow=26, window_fast=12, window_sign=9)
    data['macd'] = macd_indicator.macd()
    data['macd_signal'] = macd_indicator.macd_signal()

    # ltimos valores de MACD y seal
    macd = data['macd']
    macd_signal = data['macd_signal']

    data["MACD_TF"] = False

    for i in range(1, len(data)):
        # Si el MACD cruza hacia arriba la lnea de seal en este punto de datos
        if macd.iloc[i] > macd_signal.iloc[i] and macd.iloc[i - 1] <= macd_signal.iloc[i - 1]:
            data["MACD_TF"].iloc[i] = True

    return data
```

```python
macd_buy(data)
```

```python
def ema_cross_buy(data):
    # Crear instancia del indicador EMA con perodos de 13 y 48
    ema_13_indicator = ta.trend.EMAIndicator(close=data['Close'], window=13)
    ema_48_indicator = ta.trend.EMAIndicator(close=data['Close'], window=48)
    
    # Calcular las EMAs
    ema_13 = ema_13_indicator.ema_indicator()
    ema_48 = ema_48_indicator.ema_indicator()

    # Inicializar la seal de cruce a False
    data["ema_cross_TF"] = False

    # Iterar sobre los ndices desde el segundo valor
    for i in range(1, len(data)):
        # Si hay un cruce alcista de la EMA de 13 sobre la EMA de 48
        if ema_13.iloc[i] > ema_48.iloc[i] and ema_13.iloc[i - 1] <= ema_48.iloc[i - 1]:
            data["ema_cross_TF"].iloc[i] = True

    return data
```

```python
ema_cross_buy(data)
```

```python
def ema_200_buy(data):
    # Inicializar el indicador EMA con una ventana de 200 perodos
    ema_indicator = ta.trend.EMAIndicator(close=data['Close'], window=200)
    
    # Calcular la EMA
    ema_200 = ema_indicator.ema_indicator()
    
    # Inicializar la columna de seal de compra
    data["ema_200_TF"] = False

    # Iterar sobre los ndices desde el primer valor
    for i in range(len(data)):
        # Obtener el valor actual de la EMA de 200
        ema_200_value = ema_200.iloc[i]

        # Obtener el ltimo precio de cierre
        ultimo_precio_cierre = data['Close'].iloc[i]

        # Si el precio est por encima de la EMA de 200, establecer la seal de compra en True
        if ultimo_precio_cierre > ema_200_value:
            data.at[i, "ema_200_TF"] = True

    return data
```

```python
ema_200_buy(data)
```
